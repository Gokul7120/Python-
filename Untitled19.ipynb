```python
 import pandas as pd
 import numpy as np
 # Create a sample DataFrame
 np.random.seed(42)
 data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank','Grace', 'Hannah', 'Ivy', 'Jack'],
    'Age': np.random.randint(18, 60, size=10),
    'Gender': ['F', 'M', 'M', 'M', 'F', 'M', 'F', 'F', 'F', 'M'],
    'Salary': np.random.randint(30000, 100000, size=10),
    'Department': ['HR', 'IT', 'Finance', 'HR', 'Finance', 'IT', 'HR','Finance', 'IT', 'HR'],
    'Join_Date': pd.date_range(start='2020-01-01', periods=10,freq='M'),
    'Rating': np.random.uniform(1, 5, size=10).round(1)
 }
 df = pd.DataFrame(data)
 df
```
